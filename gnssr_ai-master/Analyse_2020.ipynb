```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset/sm_known_2020_9.0km.csv')
```

```python
# Correlation between columns in df
correlation_matrix = df.corr()
```

```python
# Display the correlation matrix in a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', annot_kws={"size": 5})
plt.savefig('figures/correlation_matrix_2020.svg', bbox_inches='tight')
```

```python
# Plot the distribution of the variable smap_sm_d
sns.distplot(df.smap_sm_d, kde=True, bins=150)  # kde=False removes the kernel density estimate
plt.savefig('figures/kde_2020_smap_sm_d.svg')
```

```python
df.boxplot(column=['smap_vwc', 'sp_inc_angle', 'smap_sm_d', 'reflectivity'])
plt.show()
```

```python
for c in ['smap_vwc', 'sp_inc_angle', 'smap_sm_d', 'pp5_3x3', 'reflectivity']:
    sns.boxplot(x='smap_vwc', data=df)
    plt.savefig(f'figures/{c}_boxplot.png')
    plt.clf()
```

```python
plt.figure(figsize=(8, 6))
ax = sns.countplot(x='lccs_class', data=df)
ax.set_yscale('log')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
plt.savefig('figures/lccs_class_distribution_log_scale.svg')
```
